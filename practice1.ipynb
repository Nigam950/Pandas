### Day-1 for learning Pandas

```python
import pandas as pd
```

#### Adding dataset in pandas

```python
data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Ankit Sharma', 'Priya Verma', 'Rahul Mehta', 'Neha Singh', 'Arjun Patel', 
             'Sneha Kapoor', 'Manish Gupta', 'Riya Malhotra', 'Karan Johar', 'Simran Kaur'],
    'Salary': [85000, 65000, 72000, 78000, 90000, 68000, 95000, 70000, 82000, 88000],
    'Role': ['Data Scientist', 'Software Developer', 'System Administrator', 'DevOps Engineer', 
             'Cloud Architect', 'QA Engineer', 'Machine Learning Engineer', 'Frontend Developer', 
             'Backend Developer', 'Database Administrator']
}
```

#### Printing dataset by making dataframe(df)

```python
df=pd.DataFrame(data)
df
```

#### Insertiong a new col [bonus]

```python
df['Bonus']=df['Salary']*.01
```

#### Inserting total col

```python
print(df)
df['total']=df['Salary']+df['Bonus']
print("After adding column")
print(df)
```

### Insert by using insert() method

```python
df.insert(0,'Gender','M')
df
```

### Using function head and tail

```python
print("Head part")
print(df.head())
print("Tail part")
print(df.tail())
```

### Shape and columns

```python
print("Shape of dataframe or dataset:")
print(df.shape)
print("Returns columns name:")
print(df.columns)
```

#### Describe and info method

```python
print("Describe Method")
print(df.describe())
print("Info Method")
print(df.info())
```

#### AND and OR Clause

```python
print("Using AND caluse")
And=df[(df['Salary']>80000) & (df['total']>=90000)]
print(And)
print("Using OR Clause")
Or=df[(df['Salary']>80000)| (df['total']>=90000)]
print(Or)
```
